# preProcessing

```python
df.drop('flight', axis=1, inplace=True)
df['class']=df['class'].apply(lambda x: 1 if x=='Economy' else 0)
```

```python
df.stops=pd.factorize(df.stops)[0]
```

```python
df
```

```python
df=df.join(pd.get_dummies(df.airline,prefix="airline",dtype=int))
df.drop('airline', axis=1, inplace=True)
df
```

```python
for colname in df.select_dtypes("object"):
  df=df.join(pd.get_dummies(df[colname],prefix=colname[:3],dtype=int))
  df.drop(colname, axis=1, inplace=True)
df
```

# Training regression Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X = df.copy()
y = X.pop("price")
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
reg=RandomForestRegressor(n_jobs=-1)
reg.fit(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error,r2_score
import math
y_pred=reg.predict(X_test)
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("RMSE:",math.sqrt(mean_squared_error(y_test,y_pred)))
print("R2:",r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
from sklearn.feature_selection import mutual_info_regression
discrete_features = X.dtypes == int
def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X, y, discrete_features)
mi_scores[::3]
```

```python
importances=dict(zip(reg.feature_names_in_,reg.feature_importances_))
sorted_importances = sorted(importances.items(), key=lambda x: x[1], reverse=True)
sorted_importances
```

```python
plt.figure(figsize=(10, 6))
plt.bar([x[0] for x in sorted_importances[:5]], [x[1] for x in sorted_importances[:5]])
plt.show()
#newlist = [expression for item in iterable if condition == True] list comprehension
```

## HyperParameter Tuning

```python
from sklearn.model_selection import GridSearchCV
reg=RandomForestRegressor(n_jobs=-1)
param_grid={
    'n_estimators':[100,200,300],
    'max_depth':[None,10,20,30],
    'min_samples_split':[2,5,10],
    'min_samples_leaf':[1,2,4],
    'max_features':['auto','sqrt']
}
grid_search=GridSearchCV(reg,param_grid,cv=5)
grid_search.fit(X_train,y_train)
best_params=grid_search.best_params_
print(best_params)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
param_dist = {
    'n_estimators':randint(100,300),
    'max_depth':randint(10,30),
    'min_samples_split':randint(2,10),
    'min_samples_leaf':randint(1,4),
    'max_features':[1.0,"auto","sqrt"]
}
reg=RandomForestRegressor(n_jobs=-1)

random_search=RandomizedSearchCV(estimator=reg,param_distributions=param_dist,cv=3,n_iter=2,scoring='neg_mean_squared_error',verbose=2,random_state=10,n_jobs=-1)
random_search.fit(X_train,y_train)
best_params=random_search.best_estimator_
print(best_params)
```

```python
best_params.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error,r2_score
import math
y_pred=best_params.predict(X_test)
print("MAE:",mean_absolute_error(y_test,y_pred))
print("MSE:",mean_squared_error(y_test,y_pred))
print("RMSE:",math.sqrt(mean_squared_error(y_test,y_pred)))
print("R2:",r2_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred)
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
d1={'social media followers':[1000,5124,None,456212,7845,45454,None,567],'Sold Out':[1,0,1,1,1,0,1,0]}
```

```python
df=pd.DataFrame(d1)
df
```

```python
x=df[["social media followers"]]
x
```

```python
y=df[["Sold Out"]]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
imputer=SimpleImputer(strategy='mean')
scaler=StandardScaler()
lr=LogisticRegression()
```

```python
pipe1=make_pipeline(imputer,scaler,lr)
```

```python
pipe1.fit(x_train,y_train)
```

```python
pipe1.score(x_train,y_train)
```

```python
d={'genre':['R','M','M','B',np.nan,'B','M','R'],'social media followers':[1000,5124,None,456212,7845,45454,None,567],'Sold Out':[1,0,1,1,1,0,1,0]}
df=pd.DataFrame(d)
df
```

```python
x=df.iloc[:,:2]
x
```

```python
y=df.iloc[:,2:]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=25)
```

```python
num_cols=["social media followers"]
cat_cols=["genre"]
```

```python
from sklearn.pipeline import Pipeline
```

```python
num_pip=Pipeline(

    steps=[('simpleImputer',SimpleImputer(strategy='mean')),
          ('scaler',StandardScaler())
          ])
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_pip=Pipeline(
    steps= [
    ('simpleImputer',SimpleImputer(strategy='most_frequent')),
    ('oneHotEncoder',OneHotEncoder())
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
preprocessor=ColumnTransformer(
    transformers=[
        ('num',num_pip,num_cols),
        ('cat',cat_pip,cat_cols)
    ],
    remainder='drop',
    n_jobs=-1
)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
```

```python
pipeFin=make_pipeline(preprocessor,dtc)
```

```python
pipeFin.fit(x_train,y_train)
```

```python
pipeFin.score(x_train,y_train)
```

```python
#save pipeline
import joblib
joblib.dump(pipeFin,'pipeFin.joblib')
```
